# 불러올 파일이 xlsx 형식일 경우

```python
import openpyxl
nu_xlsx = "./xlsx/nutrition_dataset.xlsx"
nux = pd.read_excel(nu_xlsx, dtype={'분류': str})
nux.head()
```

```python
_column_name = ['food_name', 'serving_size_g', 'energy_kcal', 'carbohydrate_g', 'total_sugars_g', 'fat_g', 'protein_g', 'calcium_mg', 'phosph_mg', 'sodium_mg', 'potassium_mg', 'magnesium_mg', 'iron_mg', 'zinc_mg', 'cholesterol_mg', 'transfat_g', 'class']
nux.columns = _column_name
```

# 불러올 파일이 csv 형식일 경우

```python
nu_csv = "./csv/nutritionDB.csv"
nuc = pd.read_csv(nu_csv, encoding='utf-8')
nuc.head(1)
```

```python
# DB군 카테고리 확인
nuc['DB군'].unique()
```

```python
# DB군 컬럼 = '음식'만 추출
meal = nuc[nuc['DB군'] == '음식']
meal['식품명'].__len__()
```

```python
# 상용제품 컬럼 = '품목대표'만 추출
standard = meal[meal['상용제품'] == '품목대표']
standard.__len__()
```

```python
# 식품명 문자열 다듬기 (공백 제거, 괄호 내용 통째로 제거)
def checkstring(df, col):
    for index, row in df.iterrows():
        value = row[col].strip()
        if '(' in value:
            df[col][index] = value[:value.index('(')]
        else:
            df[col][index] = value

checkstring(standard,'식품명')
```

```python
# 식품명 중복 체크 (모든 중복된 값이 카운트 됨)
standard.duplicated(['식품명'], keep=False).sum()
```

```python
# 중복되는 식품명은 마지막 값만 남기고 모두 삭제
standard.drop_duplicates(['식품명'], keep='last', inplace=True)
# 다시 식품명 중복 체크
standard.duplicated(['식품명'], keep=False).sum()
```

```python
standard.info()
'''
825개 레코드 -> index 번호 정리 필요
235개 컬럼 -> 정리 필요
'''
```

```python
# null 확인
standard.isnull().sum()
standard.eq('0').sum()
standard.eq(0).sum()
standard.eq('-').sum()
# null은 없지만 '-'값(사실상 null)은 많다.
```

```python
# null('-') 개수가 300개 초과인 컬럼 삭제
drop_col = standard.eq('-').sum()[standard.eq('-').sum() > 300].index.tolist()
standard.drop(columns=drop_col, inplace=True)
```

```python
# 컬럼명 확인
standard.columns
```

```python
# 1회제공량 vs 회분 ?
# standard[['식품명','1회제공량(g)', '회분(g)']]
```

```python
# 업무상 필요 없는 컬럼 삭제
standard.drop(columns=['NO', 'SAMPLE_ID', 'DB군', '상용제품', '연도', '지역 / 제조사', '채취시기', '식품대분류', '식품상세분류', '내용량_단위', '회분(g)'], inplace=True)
```

```python
# 컬럼명 변경
standard.rename(columns={'1회제공량':'1회제공량(g)',
                        '레티놀(㎍)':'비타민 A(㎍)',
                        '나이아신(㎎)':'비타민 B3(㎎)',
                        '엽산(DFE)(㎍)':'비타민 B9(㎍)'}, inplace=True)
# 비타민 E(mg) = 토코페롤+토코트리에놀
standard['비타민 E(㎎)'] = (round(standard['토코페롤(㎎)'].astype('float') + standard['토코트리에놀(㎎)'].astype('float'),2)).astype('string')
standard.drop(columns=['토코페롤(㎎)', '토코트리에놀(㎎)'], inplace=True)
```

```python
# 인덱스 리셋
standard.reset_index(drop=True, inplace=True)
```
