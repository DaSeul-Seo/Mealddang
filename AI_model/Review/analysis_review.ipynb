```python
import ast

# 문자열 형식의 리스트를 파이썬 리스트로 변환하는 함수
def convert_to_list(x):
    return [ast.literal_eval(token) for token in x]

# 각 음식점 별로 토큰들을 리스트로 저장
grouped_reviews = review.groupby('businessName')['tokenized'].apply(lambda x: convert_to_list(x))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.python.client import device_lib
from tensorflow.keras.models import load_model
```

```python
from eunjeon import Mecab
mecab = Mecab()
```

```python
tokenizer = Tokenizer()
```

```python
max_len = 80
```

```python
import re
```

```python
data = pd.read_csv('./data/tokenized_data.csv', encoding_errors='ignore', low_memory=False)
```

```python
loaded_model = load_model('./model/final_model2.h5')
```

```python
def sentiment_predict(new_sentence):
  new_sentence = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','', new_sentence)
  new_sentence = mecab.morphs(new_sentence)
  new_sentence = [word for word in new_sentence if not word in stopwords]
  encoded = tokenizer.texts_to_sequences([new_sentence])
  pad_new = pad_sequences(encoded, maxlen = max_len)

  # loaded_model.predict()의 반환값을 배열로 받음
  score = loaded_model.predict(pad_new)
  # 배열에서 첫 번째 요소를 추출하여 스칼라로 변환
  score = float(score)  # 스칼라 값으로 변환
  
  if(score > 0.5):
    print("{:.2f}% 확률로 긍정 리뷰입니다.".format(score * 100))
    return 1
  else:
    print("{:.2f}% 확률로 부정 리뷰입니다.".format((1 - score) * 100))
    return 0
```

```python
for index, row in data.iterrows():
    sp = sentiment_predict(row['body'])
    print(sp)
    if sp == 1:
        data.at[index, 'new_label'] = '1'
    elif sp == -1:
        data.at[index, 'new_label'] = '-1'
    else:
        data.at[index, 'new_label'] = '0'

    # mecab.nouns 함수를 호출하기 전에 body가 str 타입인지 확인
    if isinstance(row['body'], str):
        data.at[index, 'tokenized_comments'] = mecab.nouns(row['body'])
```

### 긍정 부정 상관관계 확인

```python
from collections import Counter
```

```python
data = pd.read_csv('./data/tokenized_data_4.csv', encoding_errors='ignore')
```

```python
data = data.dropna(subset=['rating'])
```

```python
data['label']
```

```python
# 모든 리뷰에서 동사 토큰 추출
verbs = []
for index, row in data.iterrows():
    if isinstance(row['body'], str):
        tokens = mecab.pos(row['body'])
        verbs.extend([token[0] for token in tokens if token[1].startswith('VV')])  # VV는 동사 태그입니다.

# 동사 토큰을 긍정과 부정 리뷰로 분리
positive_verbs = []
negative_verbs = []
for index, row in data.iterrows():
    if isinstance(row['body'], str):
        tokens = mecab.pos(row['body'])
        verb_list = [token[0] for token in tokens if token[1].startswith('VV')]
        if row['label'] == '1':
            positive_verbs.extend(verb_list)
        else:
            negative_verbs.extend(verb_list)

# 동사 토큰의 빈도 분석
positive_verb_counts = Counter(positive_verbs)
negative_verb_counts = Counter(negative_verbs)

# 결과 출력
print("긍정 리뷰에서 가장 많이 나온 동사 Top 10:")
print(positive_verb_counts.most_common(10))
print("\n부정 리뷰에서 가장 많이 나온 동사 Top 10:")
print(negative_verb_counts.most_common(10))
```
