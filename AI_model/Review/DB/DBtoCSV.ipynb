# DB 연결

```python
import pandas as pd
from sqlalchemy import create_engine

# MySQL 서버 연결 정보 설정
host = '15.164.171.183'
user = 'mealddang'
password = 'test1234'
database = 'mealddang'

# MySQL에 연결하는 Connection String 생성
connection_string = f'mysql+pymysql://{user}:{password}@{host}/{database}'

# 엔진 생성
engine = create_engine(connection_string)
```

# 쿼리

```python
# 쿼리 실행 및 결과 DataFrame으로 로드 Naver
query = "SELECT * FROM reviews_naver;"
df = pd.read_sql(query, engine)
```

# 데이터프레임 및 list로 저장

```python
# 데이터프레임 생성
data = []

for i in range(len(df)):
    row = [
        df.iloc[i]['review_id'],
        df.iloc[i]['rating'],
        df.iloc[i]['body'],
        df.iloc[i]['thumbnail'],
        df.iloc[i]['tags'],
        df.iloc[i]['status'],
        df.iloc[i]['visitCount'],
        df.iloc[i]['viewCount'],
        df.iloc[i]['visited'],
        df.iloc[i]['created'],
        df.iloc[i]['originType']
    ]

    if eval(df.iloc[i]['item']) is None:
        row.append("nan")
        row.append("nan")
        row.append("nan")
        row.append("nan")
    else:
        item_info = eval(df.iloc[i]['item'])
        row.append(item_info['name'])
        row.append(item_info['code'])
        row.append(item_info['options'])
        row.append(item_info['__typename'])

    

    if eval(df.iloc[i]['media']) == []:
        row.extend(["nan", "nan", "nan", "nan", "nan", "nan", "nan", "nan"])
    else:
        media_info = eval(df.iloc[i]['media'])[0]
        row.extend([
            media_info['type'],
            media_info['thumbnail'],
            media_info['thumbnailRatio'],
            media_info['class'],
            media_info['videoId'],
            media_info['videoOriginSource'],
            media_info['trailerUrl'],
            media_info['__typename']
        ])

    reply_info = eval(df.iloc[i]['reply'])
    row.extend([
        reply_info['editUrl'],
        reply_info['body'],
        reply_info['editedBy'],
        reply_info['created'],
        reply_info['date'],
        reply_info['replyTitle'],
        reply_info['isReported'],
        reply_info['isSuspended'],
        reply_info['__typename']
    ])

    has_viewer_reacted_info = eval(df.iloc[i]['hasViewerReacted'])
    row.extend([
        has_viewer_reacted_info['id'],
        has_viewer_reacted_info['reacted'],
        has_viewer_reacted_info['__typename']
    ])

    if eval(df.iloc[i]['votedKeywords']) == []:
        row.extend(["nan", "nan", "nan", "nan", "nan"])
    else:
        voted_keywords_info = eval(df.iloc[i]['votedKeywords'])[0]
        row.extend([
            voted_keywords_info['code'],
            voted_keywords_info['iconUrl'],
            voted_keywords_info['iconCode'],
            voted_keywords_info['displayName'],
            voted_keywords_info['__typename']
        ])

    data.append(row)
```
