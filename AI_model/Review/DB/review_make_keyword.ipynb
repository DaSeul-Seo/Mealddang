```python
import numpy as np
import re
```

```python
total_data = df.copy()

for i, body in total_data['body'].items():
    if isinstance(body, str):
        total_data.at[i, 'body'] = re.sub(r'[\U00010000-\U0010ffff]', '', body)
    else:
        total_data.at[i, 'body'] = np.nan
total_data = total_data.dropna(subset=['body'])
total_data.drop_duplicates(subset=['body'], inplace=True)
total_data['body'] = total_data['body'].apply(lambda x: re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣\s]', '', str(x)))
total_data['body'] = total_data['body'].fillna('')
total_data['body'] = total_data['body'].replace('', np.nan)
total_data['body'] = total_data['body'].str.replace('\n', '')

# 문자열에 strip() 메소드 적용
total_data['body'] = total_data['body'].str.strip()
total_data['body'].replace('', np.nan, inplace=True) 
total_data = total_data.dropna(subset=['body'])
total_data['body'] = total_data['body'].astype(str)
```

```python
def tokenize_text(text):
    try:
        kkma = Kkma()
        nouns = kkma.nouns(text)
        return nouns
    except Exception as e:
        print(f"에러 발생: {e}")
        return []  # 혹은 다른 적절한 처리를 수행할 수 있습니다.
```

```python
tqdm.pandas()
total_data['nouns'] = total_data['body'].progress_apply(tokenize_text)
```

```python
# total_data가 DataFrame이라고 가정하고 진행합니다.
grouped_data = total_data.groupby(['restaurant_id', 'businessName'])['nouns'].agg(sum)

# 결과 출력
print(grouped_data)
```

```python
grouped_data_df = grouped_data.reset_index()

# 결과 출력
print(grouped_data_df)
```

```python
grouped_data_df['nouns'] = grouped_data_df['nouns'].apply(lambda x: list(set(x)))

# 결과 출력
print(grouped_data_df)
```

```python
len(grouped_data_df.iloc[0]['nouns'])
```

```python
# grouped_data_df에서 'restaurant_id'와 'nouns' 열만 선택하여 새로운 DataFrame 생성
keyword_df = grouped_data_df[['restaurant_id', 'nouns']].copy()

# 'keyword' 열 생성 및 초기화
keyword_df['keyword'] = ''

# 'nouns' 열의 값을 'keyword' 열에 하나씩 저장
for index, row in keyword_df.iterrows():
    keyword_df.at[index, 'keyword'] = row['nouns']

# 결과 출력
print(keyword_df)
```
